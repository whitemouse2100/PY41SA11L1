```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y)#chưa in thực tế. Chỉ in khi gọi lệnh show
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1,2,1) 
plt.plot(x,y,'r')
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplots(3,2)
```

```python
fig=plt.figure()
#left,bottom,width,height
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.15,0.4,0.3])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("LARGER PLOT")
axes2.plot(y,x)
axes2.set_title("SMALLER PLOT")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
```

```python
#Figure Size and DPI
```

```python
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axe=plt.subplots(2,1,figsize=(8,2))
axe[0].plot(x,y)
axe[1].plot(x,y)
plt.tight_layout()
```

```python
fig.savefig("mypicture1.png",dpi=200)
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
fig=plt.figure()
ax=fig.add_axes(([0,0,1,1]))
ax.plot(x,x**2,label="X Squared")
ax.plot(x,x**3,label="X Cubed")
#ax.legend(loc=10)
ax.legend(loc=(0.1,0.1))
```

```python
#Plot Appearance
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.plot(x,y,color='green')
#linestyle= -.,:,steps,--,-
#ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=0.5,linestyle='-.') #RGB Hex code
ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=0.5,linestyle="-",marker="o",markersize=10,
        markerfacecolor="yellow",markeredgewidth=3,markeredgecolor="green"
       )
```
